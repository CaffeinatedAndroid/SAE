```python
import pandas as pd

df = pd.read_csv("Week2/energy_cleaned.csv")
df.head()

#Renewable has unusual name
```

```python
high_renew = df[df["%Renewable"] > 50]
print("Countries with >50% renewable energy:", len(high_renew))

#2 countires are mostly renewable, 1 country is at the tipping point
#the remainder are yet to cross 50% meaning they are majority petroleum based
```

```python
region_avg = df.groupby("Region")["%Renewable"].mean()
print(region_avg)

#2 countires are mostly renewable, 1 country is at the tipping point
#the remainder are yet to cross 50% meaning they are majority petroleum based
```

```python
region_avg.plot(kind='bar')

#South America has the highest average renewable use at 95%
#followed by Asia at 67% renewables
```

```python
import matplotlib.pyplot as plt

df["%Renewable"].plot(kind="hist", bins=8, edgecolor="black")
plt.title("Histogram of Renewable Energy Use")
plt.xlabel("% Renewable")
plt.grid(True)
plt.show()

#
#Changing the bins changes the visual collumn widths
```

```python
df["%Renewable"].plot(kind="density", linewidth=2)
plt.title("Density Plot of Renewable %")
plt.xlabel("% Renewable")
plt.grid(True)
plt.show()
```

```python
for region, group in df.groupby("Region"):
    if len(group) >= 2:
        group["%Renewable"].plot(kind="density", label=region)

plt.title("Renewable Energy by Region (KDE)")
plt.xlabel("% Renewable")
plt.legend()
plt.grid(True)
plt.show()
```
